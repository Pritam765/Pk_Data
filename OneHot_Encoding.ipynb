```python
import pandas as pd
```

```python
df= pd.DataFrame({'colors':['red','green','blue','green','blue','red']})
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['colors']])
```

```python
encoded=pd.DataFrame(encoder.fit_transform(df[['colors']]).toarray(),columns=encoder.get_feature_names_out())
encoded
```

```python
pd.concat([df,encoded],axis=1)
```

Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({'size':['small','medium','large','medium','small','large']})
df
```

```python
# create an instance
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

Target Guided ordinal Encoding

```python
df= pd.DataFrame({'city':['london','paris','newyork','germany','london','paris'],
                  'Price':[200,100,300,500,600,400]})
df
```

```python
mean_price=df.groupby('city')['Price'].mean().to_dict()
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
data=sns.load_dataset('tips')
data
```

```python
df=data[['total_bill','day']]
df
```

```python
mean_bill=df.groupby('day')['total_bill'].mean().to_dict()
mean_bill
```

```python
df['day_encoded']=df['day'].map(mean_bill)
```
