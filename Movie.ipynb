```python
#Movie having rating above >9

highest_rated_movies=movie[movie['imdb_score']>9]
print(highest_rated_movies[['title','imdb_score']])
```

```python
def categorize_score(score):
    if score > 8:
        return 'Best'
    elif score > 7:
        return 'Better'
    elif score > 6:
        return 'Good'
    else:
        return 'Boaring'
    
movie['Category'] = movie['imdb_score'].apply(categorize_score)

print(movie[['title','imdb_score','Category']])    
    
```

```python
counts= movie['type'].value_counts()
counts
```

```python
percentages = (counts / len(movie['type'])) * 100
fig1 = px.pie(names=percentages.index, values=percentages.values, title='Type Distribution')
fig1.show()
```
